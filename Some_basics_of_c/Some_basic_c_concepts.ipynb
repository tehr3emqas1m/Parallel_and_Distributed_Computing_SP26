# C Structrures:
A structure (or struct) is a collection of related data items, possibly of different types, grouped together under a single name.

# Basic Syntax

```python
"""
struct structure_name {
    data_type member1;
    data_type member2;
    // ... more members
};

"""
```

# Example 1: Basic Student Structure
Let's start with something simple - storing student information:



```python
%%bash
cat <<EOF > program.c
#include <stdio.h>
#include <string.h>

// Define the structure
struct Student {
    char name[50];
    int age;
    float gpa;
    int id;
};

int main() {
    // Declare a structure variable
    struct Student student1;
    
    // Access members using dot operator (.)
    strcpy(student1.name, "Alice Johnson");
    student1.age = 20;
    student1.gpa = 3.8;
    student1.id = 2024001;
    
    // Print the data
    printf("Student Information:\n");
    printf("Name: %s\n", student1.name);
    printf("Age: %d\n", student1.age);
    printf("GPA: %.2f\n", student1.gpa);
    printf("ID: %d\n", student1.id);
    
    return 0;
}

EOF
gcc program.c -o program
./program
```

# Example 2: Structure of books

```python
%%bash
cat <<EOF > program.c

#include <stdio.h>
#include <string.h>

struct Book {
    char title[100];
    char author[50];
    int year;
    float price;
};

int main() {
    // Initialize multiple structure variables
    struct Book book1 = {"The C Programming Language", "K&R", 1978, 45.99};
    struct Book book2 = {"Clean Code", "Robert Martin", 2008, 52.50};
    struct Book book3;
    
    // Assign values to book3
    strcpy(book3.title, "Python Crash Course");
    strcpy(book3.author, "Eric Matthes");
    book3.year = 2019;
    book3.price = 39.95;
    
    // Print all books
    printf("=== Book Catalog ===\n\n");
    
    printf("Book 1: %s by %s (%d) - $%.2f\n", 
           book1.title, book1.author, book1.year, book1.price);
    
    printf("Book 2: %s by %s (%d) - $%.2f\n", 
           book2.title, book2.author, book2.year, book2.price);
    
    printf("Book 3: %s by %s (%d) - $%.2f\n", 
           book3.title, book3.author, book3.year, book3.price);
    
    return 0;
}

EOF

gcc program.c -o program
./program
```

# void* in C - The Generic Pointer

void* (void pointer) is C's way of saying "pointer to something, but I don't know what type yet." It is a universal pointer that can point to any data type.



```python
%%bash
cat <<EOF > program.c

#include <stdio.h>
int main() {

int x = 42;
float y = 3.14;
char c = 'A';

void* ptr;

ptr = &x;  // OK - void* can point to int
//ptr = &y;  // OK - can point to float  
//ptr = &c;  // OK - can point to char

 printf("%d", *ptr);  // ERROR! Cannot dereference void*
   return 0;
}
EOF

gcc program.c -o program
./program
```

```python
%%bash
cat <<EOF > program.c
#include <stdio.h>

int main() {
    int x = 42;
    float y = 3.14;
    char c = 'A';
    
    void* ptr;

    // Point to int
    ptr = &x;
    printf("%d\n", *(int*)ptr);  // cast void* to int*

    // Point to float
    ptr = &y;
    printf("%.2f\n", *(float*)ptr);  // cast void* to float*

    // Point to char
    ptr = &c;
    printf("%p\n", ptr);  // cast void* to char*

    return 0;
}
EOF

gcc program.c -o program
./program
```

# void* return type 

Example: Returning Pointers to Global Variables

```python
%%bash
cat <<EOF > program.c
#include <stdio.h>

// Global variables (live forever)
int global_int = 100;
float global_float = 3.14;
char global_char = 'Z';

// Function that returns a void* to different globals based on input
void* get_global(char type) {
    if (type == 'i')
        return &global_int;      // Returns int* as void*
    else if (type == 'f')
        return &global_float;    // Returns float* as void*
    else if (type == 'c')
        return &global_char;     // Returns char* as void*
    else
        return NULL;
}

int main() {
    // Get integer global
    void* ptr = get_global('i');
    printf("Integer: %d\n", *(int*)ptr);
    
    // Get float global
    ptr = get_global('f');
    printf("Float: %f\n", *(float*)ptr);
    
    // Get char global
    ptr = get_global('c');
    printf("Char: %c\n", *(char*)ptr);
    
    return 0;
}

EOF

gcc program.c -o program
./program
```

# Stack vs Heap memory


`Stack:`

The stack is where things like:

- local variables

- function parameters

- function call information

are stored.


```python
HIGH ADDRESS
+------------------+
|      Stack       |  <- Local variables, function calls
| (grows downward) |
|                  |
|                  |
|                  |
|      Heap        |  <- malloc(), calloc(), realloc()
| (grows upward)   |
+------------------+
|  Uninitialized   |  <- BSS (global/static uninitialized)
|     Data (BSS)   |
+------------------+
|  Initialized     |  <- Data segment (global/static initialized)
|      Data        |
+------------------+
|      Code        |  <- Program instructions (text segment)
|     (Text)       |
+------------------+
LOW ADDRESS
```

# Stack visualization for the above code

```python
Step 1: Start of main (before any calls)

+-----------------+   <-- top of stack (low address)
| main frame      |
| - return addr   |
| - a (4)         |
| - b (7)         |
+-----------------+




Step 2: main calls SquareOfSum(2,3)
Push parameters (or they are in registers, but assume stack for simplicity), return address, then new frame.


+-----------------+   <-- top
| SquareOfSum     |
| - return addr   |
| - x = 2         |
| - y = 3         |
| - z (uninit)    |
+-----------------+
| main frame      |
| - return addr   |
| - a = 4         |
| - b = 7         |
+-----------------+


Step 3: SquareOfSum calls square(x+y) = square(5)
Push parameter 5, return address, new frame.

+-----------------+   <-- top
| square frame    |
| - return addr   |
| - x = 5         |
+-----------------+
| SquareOfSum     |
| - return addr   |
| - x = 2         |
| - y = 3         |
| - z (uninit)    |
+-----------------+
| main frame      |
| - return addr   |
| - a = 4         |
| - b = 7         |
+-----------------+


Step 4: square returns 25
Pop square frame. Return value passed to SquareOfSum.

+-----------------+   <-- top
| SquareOfSum     |
| - return addr   |
| - x = 2         |
| - y = 3         |
| - z = 25        |   (now initialized)
+-----------------+
| main frame      |
| - return addr   |
| - a = 4         |
| - b = 7         |
+-----------------+


Step 5: SquareOfSum returns 25
Pop SquareOfSum frame. Return value to main.

+-----------------+   <-- top
| main frame      |
| - return addr   |
| - a = 4         |
| - b = 7         |
+-----------------+



Step 6: main calls printf
Push printf frame.


+-----------------+   <-- top
| printf frame    |
| (details depend)|
+-----------------+
| main frame      |
| - return addr   |
| - a = 4         |
| - b = 7         |
+-----------------+

Step 7: printf returns
Pop printf frame.


+-----------------+   <-- top
| main frame      |
| - return addr   |
| - a = 4         |
| - b = 7         |
+-----------------+

Step 8: main returns
Pop main frame, stack empty.




```

# Stack overflow

A stack overflow in C happens when a program uses more stack memory than the stack can hold. If too much memory is used there, the program crashes.

This happens during:
- Infinite recursion
- Very large local arrays like (e.g., int arr[10000000];)




To avoid stack overflow, use:
- Avoid uncontrolled recursion
- Use the heap (malloc) for large memory
